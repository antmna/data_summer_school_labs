```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import  linear_model
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
import tensorflow as tf
```

```python
boston = load_boston()
```

```python
x_data = np.float32(boston.data)
y_data = np.float32(boston.target)

row=len(x_data)

x0=np.ones(row)

x0=x0.reshape(row,1)
x_set=np.float32(np.concatenate((x0, x_data), axis=1))
x_set=np.transpose(x_set)

W=tf.Variable(tf.random_uniform([1,14], -1.0, 1.0))


hypothesis = tf.matmul(W, x_set)
cost = tf.reduce_mean(tf.pow(hypothesis-y_data,2)) 

a = tf.Variable(0.000001)
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

init = tf.initialize_all_variables()

W_val = []
cost_val = []

sess = tf.Session()
sess.run(init)

for step in range(10000000):
    sess.run(train)
    if step % 100000 == 0:
        print (step, sess.run(cost), sess.run(W))

W_val = sess.run(W)

print("W의 첫번째 값은 b값이다.")
print ("b :", W_val[0,0], "W :", W_val[0,1:])
```
