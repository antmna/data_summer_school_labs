```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def difference_quotient(f,x,h):
    return (f(x+h)-f(x))/h
```

```python
from numpy import arange
x_vector=[i for i in arange(-4,4,0.01)]
def make_function(x):
    return x**4 + ( x**3) - (8*(x**2)) - (12*x)

y_vector=[make_function(x) for x in x_vector]

plt.plot(x_vector,y_vector)
plt.show()
```

```python
graient_vector=[difference_quotient(make_function,x,h=0.00001) for x in x_vector]

plt.plot(x_vector,graient_vector)
plt.show()
```

```python
#graient 함수 쓴 그래프랑 직접 미분 한 도함수 쓴 그래프 비교해봄
def comparison_function(x):
    return 4*x**3 + 3*x**2 - 16*x -12
s_y_vector = [comparison_function(i) for i in x_vector]

plt.plot(x_vector,s_y_vector)
plt.show()
```

```python
import math
print("using the gradient")
v=random.randint(-4,4)
print("starting point :",v)
tracking=[]
tolerance=0.0000001
while True:
    gradient = comparison_function(v)
    tracking.append(v)
    next_v=gradient*-0.001+v
    if (v - next_v <tolerance): 
        break
    v=next_v
    
print("minimum v :",v)
```
